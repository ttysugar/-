```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth, association_rules
```

```python
#數據集
dataset=[
    ['a','d','e'],
    ['a','b','c','e'],
    ['a','b','d','e'],
    ['a','c','d','e'],
    ['b','c','e'],
    ['b','d','e'],
    ['c','d'],
    ['a','b','c'],
    ['a','d','e'],
    ['a','b','e']
]
```

```python
#定義模型
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

#支持度
frequent_itemsets = fpgrowth(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
#置信度
association_result = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
association_result
```

```python
#計算支持度 {a}、{b, c}、{a,b,e}
support_a = frequent_itemsets[frequent_itemsets['itemsets'].astype(str) == "frozenset({'a'})"]
support_bc = frequent_itemsets[frequent_itemsets['itemsets'].astype(str) == "frozenset({'c', 'b'})"]
support_abe = frequent_itemsets[frequent_itemsets['itemsets'].astype(str) == "frozenset({'e', 'a', 'b'})"]
print (support_a)
print (support_bc)
print (support_abe)
```

```python
#計算置信度 {b, c} → {a} 、 {a} → {b, c}
confidence_bc = association_result[association_result['antecedents'].astype(str) == "frozenset({'c', 'b'})"]
confidence_bc_a = confidence_bc[confidence_bc['consequents'].astype(str) == "frozenset({'a'})"]
print (confidence_bc_a)
confidence_a = association_result[association_result['antecedents'].astype(str) == "frozenset({'a'})"]
confidence_a_bc = confidence_a[confidence_a['consequents'].astype(str) == "frozenset({'c', 'b'})"]
print (confidence_a_bc)
```
